# Web Scraping
In this jupyter file, we have performed web scraping with Beaturiful soup to gather all of the text from the articles. The scraped text is saved to a column in the dataframe and then saved to a csv file for later use. 

```python
import pandas as pd

df_articles_jan17 = pd.read_csv("Data/ArticlesJan2017.csv")
df_articles_feb17 = pd.read_csv("Data/ArticlesFeb2017.csv")
df_articles_mar17 = pd.read_csv("Data/ArticlesMarch2017.csv")
df_articles_apr17 = pd.read_csv("Data/ArticlesApril2017.csv")
df_articles_may17 = pd.read_csv("Data/ArticlesMay2017.csv")

df_articles_jan18 = pd.read_csv("Data/ArticlesJan2018.csv")
df_articles_feb18 = pd.read_csv("Data/ArticlesFeb2018.csv")
df_articles_mar18 = pd.read_csv("Data/ArticlesMarch2018.csv")
df_articles_apr18 = pd.read_csv("Data/ArticlesApril2018.csv")


articles = [df_articles_jan17, df_articles_feb17, df_articles_mar17, df_articles_apr17, df_articles_may17,
            df_articles_jan18, df_articles_feb18, df_articles_mar18, df_articles_apr18]

df_articles = pd.concat(articles, ignore_index=True, sort=False)
```

```python
df_articles.drop('abstract', axis=1, inplace=True)
```

```python
# cleaning out articles that are not from the New York Times
df_articles_nyt = df_articles[df_articles.webURL.str.contains("https://www.nytimes.com/")]
```

```python
# pip install beautifulsoup4
# Source: https://stackoverflow.com/questions/328356/extracting-text-from-html-file-using-python

import requests
from bs4 import BeautifulSoup

def url_to_text(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html)

    for script in soup(["script", "style"]):
        script.extract()    

    text = soup.get_text()

    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    text = '\n'.join(chunk for chunk in chunks if chunk)

    return text
```

```python
# applying the funstion above to a new column for text
df_articles_nyt['text'] = df_articles_nyt['webURL'].apply(url_to_text)
```

```python
# saving the dataframe to a csv file
df_articles_nyt.to_csv('ArticlesText.csv', index=False)
```
